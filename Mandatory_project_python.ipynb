```python
#1.What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

# Assuming df is DataFrame
avg_median_income = df['median_income'].mean()
print(avg_median_income)
# Plot the distribution using a histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['median_income'], kde=True, color='skyblue')
plt.title('Distribution of Median Income')
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.show()
#Ans "Looking at the graph, the median income distribution appears positively skewed,so that mean is likely higher than the median.
 # and the  median income is a  continuous data  representing the middle value within a range."

```

```python
#5.Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set.
import pandas as pd
#mean will give mean value of total_bedroom
mean_tb=df['total_bedrooms'].mean()
#now replacing na value with mean value
df['total_bedrooms'].fillna(mean_tb,inplace=True)
print(df)
```

```python
#6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.
import pandas as pd

def calculate_median_for_column(df, column_name):

    column_data = df[column_name]

    # Calculate the median using the pandas median function
    median = column_data.median()

    return median

median_age = calculate_median_for_column(df,'housing_median_age')
median_income = calculate_median_for_column(df,'median_income')
median_house_value=calculate_median_for_column(df,'median_house_value')
median_income_rounded=round(median_income,2)

print(f"The median for housing_median_age column is: {median_age}")
print(f"The median for median_income column is: {median_income_rounded}")
print(f"The median for median_house_value column is: {median_house_value}")
#housing_median_age (Median: 29.0):
```

```python
#10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small.
#If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.


import pandas as pd
import numpy as np

# Assuming you have a DataFrame named df with a column total_bedrooms
# If not, replace this with the actual DataFrame name and column name

# Create a new column 'total_bedroom_size' based on the specified conditions
df['total_bedroom_size'] = np.select(
    [df['total_bedrooms'] <= 10, (df['total_bedrooms'] > 10) & (df['total_bedrooms'] < 1000)],
    ['small', 'medium'],
    default='large'
)

# Display the updated DataFrame
print(df)
```
